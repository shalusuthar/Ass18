Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

Ans. DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object.

The main distinction between the two is this. For a single-column DataFrame, an index can be optional, but a Series has to have an index defined. A single-column DataFrame comprises a single column with a label while a Series lacks any columns or column labels.
